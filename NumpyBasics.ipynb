NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

```python
np.random.seed(60)
g = np.random.randint(10, size=6)
g
```

```python
# can use normal python indexing with numpy arrays
g[0:2]
```

```python
img = np.random.randn(250, 250) # creates a 250x250 matrix filled with random numbers from a uniform gaussian/normal distribution of mean 0 and variance 1
# the curve of the distribution (prolly) looks like a bell with peak at x = 0 
img
```

```python
plt.imshow(img)
```

```python
x = np.sin(img) # applies sine function on each element of the array
x
```

```python
plt.imshow(x)
```

```python
np.sum(x) # sum of all elements in the array
```

```python
np.max(x) # maximum element of the array
```

```python
np.min(x) # minimum element of the array
```

```python
np.var(x) # variance (sigma) of the array
```

```python
np.mean(x) # mean/avg of the array
```

```python
np.std(x) # standard deviation of the array
```

```python
np.prod(x) # product of all elements
```

```python
np.argmin(x) # index of smallest element
```

```python
np.argmax(x) # index of largest element
```

```python
w = np.array([1,2,3,4,5])
print(w < 3) # a binary array which runs the comparision per element of the array
print(w > 3)
```

```python
mask = np.where(img > 0.5, 255, 122) # in the image if any value is greater than 0.5 replace it with 255 else replace it with 122
plt.imshow(mask)
```

```python
a_1 = np.array([1,2,4,5])
a_2 = np.array([0,1,2,3])
a_1 + a_2
```

```python
a_1 * 10
```

```python
a_1 @ a_2 # multiplies them like matrices 25 = 1*0 + 2*1 + 4*2 + 5*3 (dot product of the two vectors)
```

```python
# 2x + 3y = 10
# 5x + 7y = 20
# to be solved
A = np.array([[2, 3],[5, 7]])
B = np.array([10, 20])
x, y = np.linalg.solve(A, B)
print('x = ', round(x))
print('y = ', round(y))
```
